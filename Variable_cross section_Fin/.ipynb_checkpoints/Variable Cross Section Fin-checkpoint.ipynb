```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
l = 0.1
b1 = 10*10**(-3)
b2 = 5*10**(-3)
d = 7.5*10**(-3)
```

```python
def Varying_Parameters(z,h,k,x):
    A=[]
    p=[]
    m=[]
    for i in range(1,grid_size-1):
        if z==1:
            h_y = (b1) + (b2-b1)*i*x/(l)
            p.append(4*h_y) # Perimeter of cross section
            A.append(h_y) # Area per unit length
            m.append(h*p[-1]/(k*h_y**(2)))
        else:
            p.append(math.pi*d) # Perimeter of cross section
            A.append(math.pi*d**(2)/4) # Area per unit length
            m.append(h*p[-1]/(k*A[-1]))
    return A,p,m
```

```python
#  (1/x^2 + (b2-b1)/(2l)*2*A_i)*T_i+1 - ( m + 2/x^2)*T_i + (1/x^2 - (b2-b1)/(2l)*2*A_i)*T_i-1 = 0
```

```python
def Matrix(grid_size,x):
    Tridia_mat=[]
    for i in range(0,grid_size-2):
        row = []
        cnt=0
        if i!=0 and i!=(grid_size-2):
            for j in range(0,grid_size-2):
                if (j==i+1):
                    row.append(1.0/(x**2) + z*(b2-b1)/(l*x*A[i]))
                elif (j==i-1):
                    row.append(1.0/(x**2) - z*(b2-b1)/(l*x*A[i]))
                elif(j==i):
                    row.append(-(2.0/(x**2) + m[i]))
                else:
                    row.append(0)
        elif i==0:
            row = [0 for j in range(0,grid_size-2)]
            row[0],row[1] = -(2.0/(x**2) + m[i]),(1.0/(x**2) + z*(b2-b1)/(l*x*A[i]))
        else:
            row = [0 for j in range(0,grid_size-2)]
            row[-1],row[-2] = -(2.0/(x**2) + m[i]),(1.0/(x**2) - z*(b2-b1)/(l*x*A[i]))
        Tridia_mat.append(np.array(row))

    Tridia_mat = np.array(Tridia_mat)
    return Tridia_mat
```

```python
def boundary_conditions(x,T1,bdry_type='insulated',T_t = 0):
    B = [0 for i in range(0,grid_size-2)]
    B[0] = -(1.0/(x**2) - z*(b2-b1)/(l*x*b1))*T1
    B[-1] = T_t
    return B
```

```python
def TDMA(mat,B,grid_size,x,Ar,bdry_type='insulated'):
    for i in range(1,grid_size-2):
        mat[i][i] = mat[i][i] - (float(mat[i][i-1])/mat[i-1][i-1])*mat[i-1][i]
        B[i] = B[i] - (float(mat[i][i-1])/mat[i-1][i-1])*B[i-1]
        mat[i][i-1] = 0
    
    X=[0 for j in range(grid_size-3,-1,-1)]
#     print(A,B)
    for j in range(grid_size-3,-1,-1):
        if j!=(grid_size-3):
            X[j] = (B[j] - mat[j][j+1] * X[j + 1]) / mat[j][j]
        elif (bdry_type == 'insulated'):
#             print(A[j][j],1.0/(x**2), (b2-b1)/((4*l*x)*Ar[i]))
            X[j] =B[j]/(mat[j][j]+(1.0/(x**2) + z*(b2-b1)/(l*x*Ar[i])))
#             print(X[j])
        elif (bdry_type == 't_spec'):
            X[j] =B[j]/(mat[j][j])
        elif (bdry_type == 'conv'):
            X[j] =B[j]/(mat[j][j]+(1.0/(x**2) + z*(b2-b1)/(l*x*Ar[i]))/(1+(h*x/k)))
    return X
        
```

```python
def find_sol(grid_size,x,T1,bdry_type,A):
    Tridia_mat = Matrix(grid_size,x)
    B = boundary_conditions(x,T1,bdry_type=bdry_type,T_t=0)
#     print(Tridia_mat,B)
    X =TDMA(Tridia_mat,B,grid_size,x,A,bdry_type=bdry_type)
    return X
```
